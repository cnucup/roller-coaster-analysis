```python
def plot_ranking(dataset, dataset_name, name, park):
    if dataset_name == "wood":
        year_list = wood[(wood["Name"] == name) & (wood["Park"] == park)]["Year of Rank"].values
        rank_list = wood[(wood["Name"] == name) & (wood["Park"] == park)]["Rank"].values
        ax=plt.subplot()
        plt.plot(year_list, rank_list)
        plt.xlabel("Year")
        plt.ylabel("Ranking")
        plt.axis([min(year_list), max(year_list), max(rank_list), 0])
        plt.title(name + " Rankings")
        #ax.set_xticks()
        #ax.set_xticklabels()
        plt.show()
    else:
        year_list = steel[(steel["Name"] == name) & (steel["Park"] == park)]["Year of Rank"].values
        rank_list = steel[(steel["Name"] == name) & (steel["Park"] == park)]["Rank"].values
        ax=plt.subplot()
        plt.plot(year_list, rank_list)
        plt.xlabel("Year")
        plt.ylabel("Ranking")
        plt.axis([min(year_list), max(year_list), max(rank_list), 0])
        plt.title(name + " Rankings")
        #ax.set_xticks()
        #ax.set_xticklabels()
        plt.show()    
```

```python
def plot_two_coaster(coaster_1, coaster_2):
    dataset_name = coaster_1[1]
    
    if dataset_name == "wood":
        year_list_1 = wood[(wood["Name"] == coaster_1[2]) & (wood["Park"] == coaster_1[3])]["Year of Rank"].values
        rank_list_1 = wood[(wood["Name"] == coaster_1[2]) & (wood["Park"] == coaster_1[3])]["Rank"].values
        year_list_2 = wood[(wood["Name"] == coaster_2[2]) & (wood["Park"] == coaster_2[3])]["Year of Rank"].values
        rank_list_2 = wood[(wood["Name"] == coaster_2[2]) & (wood["Park"] == coaster_2[3])]["Rank"].values
        ax=plt.subplot()
        plt.plot(year_list_1, rank_list_1)
        plt.plot(year_list_2, rank_list_2)
        plt.xlabel("Year")
        plt.ylabel("Ranking")
        plt.axis([min(year_list_1.tolist() + year_list_2.tolist()), max(year_list_1.tolist() + year_list_2.tolist()), 
                  max(rank_list_1.tolist() + rank_list_2.tolist()), 0])
        plt.title(coaster_1[2] + " vs " + coaster_2[2] + " Rankings")
        plt.legend([coaster_1[2], coaster_2[2]])
        #ax.set_xticks()
        #ax.set_xticklabels()
        plt.show()
    else:
        year_list_1 = steel[(steel["Name"] == coaster_1[2]) & (steel["Park"] == coaster_1[3])]["Year of Rank"].values
        rank_list_1 = steel[(steel["Name"] == coaster_1[2]) & (steel["Park"] == coaster_1[3])]["Rank"].values
        year_list_2 = steel[(steel["Name"] == coaster_2[2]) & (steel["Park"] == coaster_2[3])]["Year of Rank"].values
        rank_list_2 = steel[(steel["Name"] == coaster_2[2]) & (steel["Park"] == coaster_2[3])]["Rank"].values
        ax=plt.subplot()
        plt.plot(year_list_1, rank_list_1)
        plt.plot(year_list_2, rank_list_2)
        plt.xlabel("Year")
        plt.ylabel("Ranking")
        plt.axis([min(year_list_1.tolist() + year_list_2.tolist()), max(year_list_1.tolist() + year_list_2.tolist()), 
                  max(rank_list_1.tolist() + rank_list_2.tolist()), 0])
        plt.title(coaster_1[2] + " vs " + coaster_2[2] + " Rankings")
        plt.legend([coaster_1[2], coaster_2[2]])
        #ax.set_xticks()
        #ax.set_xticklabels()
        plt.show()     
```

```python
def plot_n (n, dataset):
    partial_df = dataset[dataset.Rank <= n]
    coaster_list = set(partial_df.Name.tolist())
    ax = plt.subplot()
    year_min = 2020
    year_max = 0
    for coaster in coaster_list:
        coaster_year = partial_df[partial_df.Name == coaster]["Year of Rank"].values
        coaster_ranks = partial_df[partial_df.Name == coaster]["Rank"].values
        if max(coaster_year.tolist())>year_max:
            year_max = max(coaster_year.tolist())
        if min(coaster_year.tolist())<year_min:
            year_min = min(coaster_year.tolist())
        plt.plot(coaster_year, coaster_ranks)
    plt.axis([year_min, year_max, 6,0])
    plt.xlabel("Year")
    plt.ylabel("Ranking")
    plt.title("Top " + str(n) + " Ranking")
    plt.legend(coaster_list)
    plt.show()
```

```python
plot_n(5, wood)
```

```python
def plot_hist (dataset, column_name):
    plt.hist(dataset[column_name], bins=40)
    plt.xlabel(column_name.capitalize())
    plt.ylabel("Counts")
    plt.title("Roller Coaster " + column_name.capitalize() + " Histogram")
    plt.show()
```

```python
plot_hist(roller_coasters, "height")
```

```python
def plot_rc_bar (dataset, park_name):
    df = dataset[dataset["park"] == park_name]
    plt.bar(df["name"].tolist(), df["num_inversions"].tolist())
    plt.xlabel("Roller Coaster")
    plt.xticks(rotation=90)
    plt.ylabel("Number of Inversions")
    plt.title("Number of Inversion by Roller Coasters in " + park_name)
    plt.show()
```

```python
def operation_status(dataset):
    num_oper = dataset[dataset["status"]=="status.operating"].name.count()
    num_non_oper = dataset[dataset["status"]=="status.closed.definitely"].name.count()
    data = [num_oper, num_non_oper]
    labels = ["Operating", "Non-Operating"]
    explode = (0.1, 0)
    plt.pie(data, labels=labels, explode=explode, autopct="%0.1f%%")
    plt.axis("equal")
    plt.title("Operation Status of Roller Coasters")
    plt.show()
```

```python
operation_status(roller_coasters)
```

```python
def rc_scatter (dataset, col_name_1, col_name_2):
    plt.scatter(dataset[col_name_1], dataset[col_name_2])
    plt.xlabel(col_name_1.capitalize())
    plt.ylabel(col_name_2.capitalize())
    plt.title("Roller Coaster " +col_name_1.capitalize() + " vs " + col_name_2.capitalize())
    plt.show()
```

```python
rc_scatter(roller_coasters, "height", "speed")
```

```python
def seating_type(dataset):
    seating_df = dataset.dropna()
    seating_types = seating_df["seating_type"].unique().tolist()
    num_seating = []
    explode = []
    for item in seating_types:
        num_seating.append(dataset[dataset["seating_type"]==item].name.count())
        explode.append(0.3)
    plt.figure(figsize=(10,8))
    plt.pie(num_seating, labels=seating_types, explode=explode, autopct="%0.1f%%", pctdistance=0.75)
    plt.axis("equal")
    plt.title("Types of Seating by Roller Coasters")
    plt.show()
```

```python
seating_type(roller_coasters)
```

```python
def seating_type_analysis(dataset, column_name):
    seating_df = dataset.dropna()
    seating_types = seating_df["seating_type"].unique().tolist()
    average_data_per_seating_type = []
    num_seating = []
    for item in seating_types:
        num_seating.append(dataset[dataset["seating_type"]==item].name.count())
        average_data_per_seating_type.append(seating_df[seating_df["seating_type"]==item][column_name].mean())
    #average_data_per_seating_type.sort()
    plt.bar(seating_types, average_data_per_seating_type)
    plt.xticks(rotation=90)
    plt.xlabel("Seating Types")
    plt.ylabel(column_name.capitalize())
    plt.title("Roller Coaster Seating Type vs " + column_name.capitalize())
    plt.show()
```

```python
seating_type_analysis(roller_coasters, "height")
```

```python
def manufacturer_analysis(dataset, manufacturer):
    dataset["material_new"] = dataset["material_type"].where(dataset['material_type'] != "na", np.nan)
    partial_df = dataset.dropna()
    manufacturer_df = partial_df[partial_df["manufacturer"] == manufacturer]
    # material type analysis
    material_type = manufacturer_df["material_type"].unique().tolist()
    material_count = []
    for item in material_type:
        material_count.append(manufacturer_df[manufacturer_df["material_type"] == item].name.count())
    # ax1 = plt.subplot(2,1,1)
    fig1 = plt.figure(figsize=(8,8))
    plt.pie(material_count, labels = material_type, autopct="%0.1f%%")
    plt.axis("equal")
    plt.title("Material Type Analysis for "+manufacturer)
    
    # seating type analysis
    seating_type = manufacturer_df["seating_type"].unique().tolist()
    seating_count = []
    for item in seating_type:
        seating_count.append(manufacturer_df[manufacturer_df["seating_type"] == item].name.count())
    #ax2 = plt.subplot(2,1,2)
    fig2 = plt.figure(figsize=(8,8))
    plt.pie(seating_count, labels = seating_type, autopct="%0.1f%%")
    plt.axis("equal")
    plt.title("Seating Type Analysis for "+manufacturer)  
    
    # speed analysis
    fig3 = plt.figure(figsize=(8,8))
    plt.hist(manufacturer_df["speed"], bins=10)
    plt.xlabel("Speed")
    plt.ylabel("Count")
    plt.title("Speed Analysis for "+manufacturer)
    
    # length analysis
    fig4 = plt.figure(figsize=(8,8))
    plt.hist(manufacturer_df["length"], bins=10)
    plt.xlabel("Length")
    plt.ylabel("Count")
    plt.title("Length Analysis for "+manufacturer)
```

```python
manufacturer_analysis(roller_coasters, "Vekoma")
```
